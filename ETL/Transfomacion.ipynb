```python
import pandas as pd
```

```python
# Importamos archio electic car data aportados por Henry
Car_data = 'ElectricCarData_Clean.csv'
df= pd.read_csv(Car_data)
```

```python
df.info()
```

```python
# Revisamos contenido
df.head(2)
```

```python
# Buscamos marcas unicas de coches electricos
df['Brand'].unique()
```

```python
# Tomamos en consideración aquello que tienen un precio menor a 30K Euros

autos_menos_30k = df[df['PriceEuro'] < 30000]

# Convertimos el precio Euro a Dolares a los fines de comparar con el dataframe construidos por el equipo
autos_menos_30k['Precio Dolares'] = autos_menos_30k['PriceEuro']*1.09
```

```python
# Cargamos nuestro dataframe construido por nosotros
dfexcel = pd.read_excel('Marcas de Vehiculos mas Usados para Uber (1) (2).xlsx', sheet_name='Hoja2')
```

```python
dfexcel.head(2)
```

```python
#Creamos columna Marca y modelo
autos_menos_30k['Marca y modelo'] = autos_menos_30k['Brand']+' '+autos_menos_30k['Model']
```

```python
#Creamos Columna Tipo combustible
autos_menos_30k["Tipo Combustible"] = autos_menos_30k.apply(lambda x: "Electrico", axis=1)
```

```python
# Elejimos columnas a analizar
columnas = ['Tipo Combustible','BodyStyle','Marca y modelo','Precio Dolares','Range_Km']
agregar = autos_menos_30k[columnas]
agregar.head(2)
```

```python
# Renombramos columnas
agregar.rename(columns={'BodyStyle' : 'Tipo Vehiculo', 'Precio Dolares' : 'Valor Vehiculo','Range_Km' :'Kms x carga'},inplace=True) 
```

```python
# Creamos columnas a partir de los datos del dataset de autos electricos a los fines de comparar con nuestro dataframe construido

agregar['Amortizacion Vehiculo a 2 años'] = agregar['Valor Vehiculo'] / 24 #se divide en el estimado de 2 años de amortizacion vehiculo
agregar['Intereses Totales'] = agregar['Valor Vehiculo'] * 0.06
agregar['Intereses mensules'] = agregar['Intereses Totales'] / 24
agregar['Salario Conductor'] = agregar.apply(lambda x: 2500, axis=1)
agregar['Multas'] = agregar.apply(lambda x: 175, axis=1)
agregar['Impuestos'] = agregar['Valor Vehiculo']*0.09/12
agregar['Seguro Obligatorio'] = agregar.apply(lambda x: 150, axis=1)
agregar['Seguro Todo Riesgo'] = (150 + agregar['Valor Vehiculo']/2000)/24  # formula para sacar el seguro todo riesgo =(150+(D2/2000))/24
agregar['Kms Rec'] = agregar.apply(lambda x: 2000, axis=1)
agregar['costo por carga'] = agregar.apply(lambda x: 7.5, axis=1) #costo por carga
agregar['Kw/h'] = agregar.apply(lambda x: 0.10, axis=1)
agregar['Cargas / mes'] = agregar['Kms Rec']/agregar['Kms x carga']
agregar['Costo Electricidad'] = agregar['costo por carga']*agregar['Cargas / mes']
agregar['Llantas'] = 304/12 #calculo de costo de uso
agregar['Parqueadero'] = agregar.apply(lambda x: 90 ,axis=1)
agregar['Limpieza'] = agregar.apply(lambda x: 15 ,axis=1)
agregar['Depreciacion'] = agregar['Valor Vehiculo']/60
agregar['Repuestos'] = agregar['Valor Vehiculo']*0.01
agregar['Ganancia mensual'] = agregar.apply(lambda x: 7500 ,axis=1)
agregar['Operacion de App/Plataforma'] = agregar.apply(lambda x : 1500, axis=1)
```

```python
#Creamos Total operacion', 'Ganancia mensual', 'Total Ganancias'
sumacostos = ['Amortizacion Vehiculo a 2 años','Intereses mensules','Salario Conductor','Multas','Impuestos','Seguro Obligatorio','Seguro Todo Riesgo','Costo Electricidad','Llantas','Parqueadero','Limpieza','Depreciacion','Repuestos']
agregar['Total operacion Sin App'] = agregar[sumacostos].sum(axis=1)
agregar['Costo Total operacion'] = agregar['Total operacion Sin App']+ agregar['Operacion de App/Plataforma']
agregar['Total Ganancias'] = agregar['Ganancia mensual'] - agregar['Costo Total operacion']
```

### Procedemos a Concatener nuestros datasets

```python
df1 = pd.concat([dfexcel,agregar],axis=0)
```

```python
#Guardamos el avance
df1.to_csv('final_vehiculos_electricos.csv')
```
