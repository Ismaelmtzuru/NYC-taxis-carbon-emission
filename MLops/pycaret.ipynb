```python
from pycaret.classification import *
from pycaret.regression import *
from pycaret.regression import *
import pandas as pd
```

```python
data = pd.read_csv('dataset\\clima.csv')
```

```python
# Se elimina los datos faltantes o nulos para que no exista alguna interferencia con Pycaret
data.drop(data[data['temperature_2m_mean'].isnull()].index,inplace=True)
```

```python
data
```

```python
# Se configura en 'target' la columna 'temperature_2m_max' porque deseamos predecir la temperatura máxima
s = setup(data, target = 'temperature_2m_max', session_id = 123)
```

En [Target type -->	Regression] nos indica que el tipo de modelo será de regresión linear.

```python
# Lo usamos para comparar todos los tipos de regresión que existe
best = compare_models()
```

```python
# Indica que el mejor regresos es 'HuberRegressor'
print(best)
```

```python
# se evalua el modelo.
evaluate_model(best)
```

```python
'''Aqui nos muestra que el modelo es altamente confiable. En las dos últimas colúmnas se puede ver la columna
que se desea predecir y en 'prediction_label' se muestra la predicción'''
predict_model(best)
```

```python
predictions = predict_model(best)
predictions.head()
```

```python
'''Aquí se guarda el modelo para un uso posterior.'''
save_model(best,'prediccion_temp_max')
```
